```python
import sys
sys.path.append("../jockmktbot/")
import firebase_admin
from firebase_admin import db
import random
```

```python
cred = firebase_admin.credentials.Certificate(('./dbkey/key.json'))
firebase_admin.initialize_app(cred, {'databaseURL' : 'https://jockmktbot.firebaseio.com/'})
ref = db.reference() #root
```

```python
randomEvent = random.choice(list(dict(ref.get()).keys()))
print(f'Random event is {randomEvent}')
eventRef = ref.child(randomEvent)
eventStocks = dict(eventRef.get()).keys()
print(f'Event stocks are {eventStocks}')
```

```python
masterDict = {}
# {stock1: [(time1, data1), (time2, data2)]}
for stock in eventStocks:
    masterDict[stock] = []
    stockRef = eventRef.child(stock)
    for dataPoint in dict(stockRef.get()).values():
        masterDict[stock].append([dataPoint['time'], dataPoint])
print(f'Some random stock in masterDict {masterDict["stk_442638d7b951610e6a56d66c32a3453e"]}')
```

```python
for event in masterDict:
    masterDict[event].sort(key = lambda x: x[0])
```

```python
for data in masterDict['stk_833cf80038a52002b906ef675254f91f']:
    datapoint = data[1]
    print(f'Stock price {datapoint.get("estimated_price", None)}, Estimated fantasy points {datapoint.get("fantasy_points_projected_live", None)}, amount completed {datapoint.get("amount_completed")} {datapoint["stock"]["details"]["first_name"]} {datapoint["stock"]["details"]["last_name"]}')

```
